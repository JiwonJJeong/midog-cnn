```python
import pandas as pd
import matplotlib.pyplot as plt
from ast import literal_eval

# --- Start of Fixed Code using category_id for Table Generation ---

# 1. Map category_id directly to the classification
# We assume the user intends to use the previous mapping: {2: 'Mitotic', 1: 'Non-Mitotic'}
# Note: Since the previous request used {2: 'Non-Mitotic', 1: 'Mitotic'}, I am sticking to the mapping implied by the user's initial code, which was based on the 'labels' column, but now applied to the 'category_id'.
label_mapping = {2: 'Mitotic', 1: 'Non-Mitotic'}
df_merged['Mitotic_Class'] = df_merged['category_id'].map(label_mapping).fillna('Other')

# 2. Group by metadata and the new class, then count occurrences
# Since category_id is one per row, NO explode step is needed.
df_counts = df_merged.groupby(['Tumor', 'Scanner', 'Origin', 'Mitotic_Class']).size().unstack(fill_value=0)

# 3. Calculate Total Annotations and Ratios
# Ensure we handle cases where 'Mitotic' or 'Non-Mitotic' column might be missing after unstacking
mitotic_counts = df_counts.get('Mitotic', 0)
non_mitotic_counts = df_counts.get('Non-Mitotic', 0)

df_counts['Total Annotations'] = mitotic_counts + non_mitotic_counts

# Calculate Ratios
df_counts['Mitotic Ratio'] = mitotic_counts / df_counts['Total Annotations']
df_counts['Non-Mitotic Ratio'] = non_mitotic_counts / df_counts['Total Annotations']

# 4. Select and format the final columns
df_final_table = df_counts[[
    'Total Annotations', 
    'Mitotic', 
    'Non-Mitotic', 
    'Mitotic Ratio', 
    'Non-Mitotic Ratio'
]].copy()

# Format ratios as percentages
df_final_table['Mitotic Ratio'] = (df_final_table['Mitotic Ratio'] * 100).map('{:.2f}%'.format)
df_final_table['Non-Mitotic Ratio'] = (df_final_table['Non-Mitotic Ratio'] * 100).map('{:.2f}%'.format)

# Reset index and prepare for Matplotlib rendering
df_final_table = df_final_table.reset_index()
df_final_table = df_final_table.rename(columns={
    'Mitotic': 'Mitotic Annotations',
    'Non-Mitotic': 'Non-Mitotic Annotations'
})

# --- Matplotlib Table Rendering ---

# Set up the figure and axes
fig, ax = plt.subplots(figsize=(15, 3)) 

# Hide the axes
ax.axis('tight')
ax.axis('off')

# Prepare cell text and column headers
cell_text = df_final_table.values
col_labels = df_final_table.columns

# Create the table
table = ax.table(
    cellText=cell_text, 
    colLabels=col_labels, 
    cellLoc='center', 
    loc='center'
)

# Optional: Adjust font size for better fit
table.auto_set_font_size(False)
table.set_fontsize(8)
table.scale(1, 1.5) 

ax.set_title('Annotation Summary and Mitotic Ratios by Combination', y=1.2)
plt.tight_layout()
plt.savefig('category_id_summary_table.png')
```

```python
import pandas as pd
import numpy as np
from typing import Tuple, Optional

def check_uniform_bbox_dimensions(df: pd.DataFrame, bbox_column: str = 'bbox') -> Tuple[bool, Optional[Tuple[float, float]]]:
    """
    Tests whether all bounding boxes in the DataFrame have the exact same 
    width and height.

    Args:
        df (pd.DataFrame): The DataFrame containing the bounding box data.
        bbox_column (str): The name of the column containing the bounding box lists 
                           ([xmin, ymin, xmax, ymax]).

    Returns:
        Tuple[bool, Optional[Tuple[float, float]]]: 
            - bool: True if all dimensions are uniform, False otherwise.
            - Tuple[float, float]: The uniform (width, height) if True, or None if False.
    """
    if df.empty:
        print("Error: DataFrame is empty.")
        return False, None

    dimensions = []
    
    # 1. Calculate Width and Height for every valid bounding box
    for bbox in df[bbox_column]:
        if isinstance(bbox, list) and len(bbox) >= 4:
            # Calculate width and height directly from coordinates
            width = bbox[2] - bbox[0]  # xmax - xmin
            height = bbox[3] - bbox[1] # ymax - ymin
            dimensions.append((width, height))
        else:
            # If any entry is invalid, we treat the dataset as non-uniform
            return False, None

    if not dimensions:
        print(f"Error: No valid bounding box data found in column '{bbox_column}'.")
        return False, None

    # 2. Check for uniformity
    
    # Get the dimensions of the very first box as the reference
    reference_width, reference_height = dimensions[0]

    # Convert to a NumPy array for fast element-wise comparison
    dims_array = np.array(dimensions)
    
    # Check if all widths match the reference width AND all heights match the reference height
    all_widths_match = np.all(dims_array[:, 0] == reference_width)
    all_heights_match = np.all(dims_array[:, 1] == reference_height)

    is_uniform = all_widths_match and all_heights_match
    
    if is_uniform:
        return True, (reference_width, reference_height)
    else:
        return False, None

# --- Example Usage ---

is_uniform, uniform_dims = check_uniform_bbox_dimensions(df_merged)

if is_uniform:
    print(f"\n PASS: All bounding boxes are UNIFORM!")
    print(f"   Uniform Dimensions: Width={uniform_dims[0]:.2f}, Height={uniform_dims[1]:.2f}")
else:
    print(f"\n FAIL: Bounding box dimensions are NOT uniform.")
    print("   The dataset contains at least two boxes with different dimensions.")
```

```python
import matplotlib.pyplot as plt
import numpy as np
import torch

# --- 1. Function to Display a Batch ---

def visualize_dataloader_batch(data_loader, num_images=8, label_mapping={0: 'Mitotic', 1: 'Non-Mitotic'}):
    """
    Samples a batch from the DataLoader, converts the images back to a displayable
    format, and plots them in a grid.
    
    Args:
        data_loader (torch.utils.data.DataLoader): The DataLoader object.
        num_images (int): The maximum number of images to display from the batch.
        label_mapping (dict): A dictionary to map numeric labels (1, 2) to names.
    """
    print("Sampling and visualizing a batch from the DataLoader...")
    
    # Get one batch of data
    try:
        batch = next(iter(data_loader))
    except StopIteration:
        print("Error: DataLoader is empty.")
        return

    images = batch['image']
    labels = batch['label']
    
    # Calculate the grid size
    display_count = min(num_images, images.shape[0])
    cols = 4  # Fixed number of columns for a nice grid
    rows = (display_count + cols - 1) // cols
    
    plt.figure(figsize=(3.5 * cols, 3.5 * rows))
    plt.suptitle("DataLoader Output: Sampled Cropped Annotations", fontsize=14, y=1.02)
    
    for i in range(display_count):
        ax = plt.subplot(rows, cols, i + 1)
        
        # 1. Reverse the PyTorch transformation (Tensor to Numpy)
        image = images[i].cpu().numpy()
        
        # 2. Transpose the image from (C, H, W) to (H, W, C) for Matplotlib
        image = np.transpose(image, (1, 2, 0))
        
        # 3. Clip values to the valid range [0, 1] after conversion
        image = np.clip(image, 0, 1)

        # 4. Get the human-readable label
        label_id = labels[i].item()
        label_text = label_mapping.get(label_id, f"Label {label_id}")
        
        # Display the image and title
        ax.imshow(image)
        ax.set_title(f"Label: {label_text} (ID: {label_id})", fontsize=10)
        ax.axis("off")

    plt.tight_layout(rect=[0, 0, 1, 0.98]) # Adjust layout to make room for suptitle
    plt.show()

# --- 2. Execute Visualization ---

# Run the visualization function using your active data_loader
visualize_dataloader_batch(train_loader, num_images=8)
```

```python
import matplotlib.pyplot as plt
import numpy as np
import os
from PIL import Image
import pandas as pd
from tqdm import tqdm
from typing import Dict, Any, List, Tuple

# --- Configuration Constants (Must match preprocessing) ---
LABEL_COL = 'category_id'
LABEL_MAPPING = {1: 'Mitotic', 2: 'Non-Mitotic'}
GROUP_COLS = ['Origin', 'Tumor', 'Scanner']
# --------------------------------------------------------

def compute_average_patches_grouped(
    df: pd.DataFrame, 
    patch_dir: str, 
    group_cols: list = GROUP_COLS, 
    label_col: str = LABEL_COL,
    label_mapping: Dict[int, str] = LABEL_MAPPING,
    max_samples: int = 500
) -> Dict[str, Tuple[np.ndarray, int]]:
    """
    Computes and STORES the pixel-wise average image for each (Group + Class) combination.
    
    Returns:
        Dict[str, Tuple[np.ndarray, int]]: A dictionary mapping the composite key 
        ('Origin | Tumor | Scanner | ClassName') to (mean_image_array, sample_count).
    """
    results_store = {}
    
    # 1. Create a composite key for grouping
    df['Composite_Group'] = df[group_cols].apply(lambda row: ' | '.join(row.values.astype(str)), axis=1)
    unique_groups = df['Composite_Group'].unique()
    
    classes = sorted(label_mapping.keys())
    
    print(f"Computing averages for {len(unique_groups)} unique combinations...")
    tqdm_groups = tqdm(enumerate(unique_groups), total=len(unique_groups), desc="Processing Groups")

    for row_idx, group_key in tqdm_groups:
        df_group = df[df['Composite_Group'] == group_key]
        
        for class_id in classes:
            class_name = label_mapping[class_id]
            df_subset = df_group[df_group[label_col] == class_id]
            
            # Key to store the result: Origin|Tumor|Scanner|Class
            store_key = f"{group_key} | {class_name}"
            
            if len(df_subset) == 0:
                results_store[store_key] = (None, 0)
                continue
            
            # --- Load Images and Compute Average ---
            if len(df_subset) > max_samples:
                df_subset = df_subset.sample(n=max_samples, random_state=42)
            
            patch_ids = df_subset['patch_id'].values
            
            # Attempt to determine image shape from the first patch
            try:
                first_img_path = os.path.join(patch_dir, f"{patch_ids[0]}.png")
                with Image.open(first_img_path) as img:
                    img_shape = (img.size[1], img.size[0], 3) # (H, W, C)
            except (FileNotFoundError, IndexError):
                results_store[store_key] = (None, 0)
                continue

            batch_images = []
            for pid in patch_ids:
                path = os.path.join(patch_dir, f"{pid}.png")
                try:
                    img = np.array(Image.open(path).convert('RGB'), dtype=np.float32) / 255.0
                    if img.shape == img_shape: 
                        batch_images.append(img)
                except Exception:
                    continue
            
            if batch_images:
                mean_image = np.mean(np.stack(batch_images), axis=0)
                results_store[store_key] = (mean_image, len(batch_images))
            else:
                results_store[store_key] = (None, 0)

    print("\n All averages computed and stored.")
    return results_store

# --- EXECUTION FLOW ---

PATCH_DIR = 'cropped_images/' 
df_processed = df_master

stored_averages = compute_average_patches_grouped(df_processed, PATCH_DIR)
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from typing import Dict, Tuple

def plot_scanner_comparison(
    results_store: Dict[str, Tuple[np.ndarray, int]],
    target_tumor: str = 'human breast cancer',
    target_origin: str = 'UMC Utrecht'
):
    """
    Plots a simple row comparison of average images for a specific Tumor/Origin 
    pair across all available Scanners, separated by Mitotic/Non-Mitotic class.
    
    Args:
        results_store (Dict): Dictionary containing keys like 'Origin | Tumor | Scanner | Class'.
        target_tumor (str): The specific tumor type to filter for.
        target_origin (str): The specific origin to filter for.
    """
    
    # --- 1. Filter and Prepare Data ---
    
    # Key structure: 'Origin | Tumor | Scanner | Class'
    target_prefix = f"{target_origin} | {target_tumor} | "
    
    filtered_data = {}
    for key, (image, count) in results_store.items():
        if key.startswith(target_prefix) and image is not None and count > 0:
            # Extract Scanner and Class from the key
            parts = key.split(' | ')
            scanner = parts[2]
            class_type = parts[3]
            
            if scanner not in filtered_data:
                filtered_data[scanner] = {}
            filtered_data[scanner][class_type] = {'Image': image, 'Count': count}
            
    if not filtered_data:
        print(f"Error: No data found for Tumor: '{target_tumor}' and Origin: '{target_origin}'.")
        return

    # --- 2. Setup Plot Dimensions ---

    unique_scanners = sorted(filtered_data.keys())
    n_scanners = len(unique_scanners)
    
    # We will have 2 rows (Mitotic, Non-Mitotic) and N columns (Scanners)
    fig_width = n_scanners * 3.5
    fig_height = 2.5 * 2 + 1 # 2 rows, plus space for titles
    
    fig, axes = plt.subplots(
        2, n_scanners, 
        figsize=(fig_width, fig_height), 
        squeeze=False, # Ensure axes is 2D even for n_scanners=1
        gridspec_kw={'wspace': 0.1, 'hspace': 0.4}
    )
    
    plt.suptitle(f"Scanner Comparison for: {target_tumor} at {target_origin}", 
                 fontsize=16, y=0.98)

    # --- 3. Plotting Loop ---
    
    class_map = {'Mitotic': 0, 'Non-Mitotic': 1} # Row mapping
    
    for c, scanner in enumerate(unique_scanners):
        scanner_data = filtered_data[scanner]
        
        # --- Column Title (Scanner Name) ---
        axes[0, c].set_title(scanner, fontsize=12, fontweight='bold', pad=10)
        
        for class_type, row_idx in class_map.items():
            ax = axes[row_idx, c]
            item = scanner_data.get(class_type)
            
            if item:
                image = item['Image']
                count = item['Count']
                
                ax.imshow(image)
                
                # Label below the image
                color = 'darkred' if class_type == 'Mitotic' else 'darkblue'
                label = f"{class_type[0]} ({count})"
                ax.text(0.5, 1.15, label, transform=ax.transAxes, 
                        ha='center', fontsize=10, color=color, va='bottom')
            else:
                ax.text(0.5, 0.5, "No Data", ha='center', fontsize=8)

            # --- Row Label (Left Side) ---
            if c == 0:
                ax.set_ylabel(class_type, fontsize=12, fontweight='bold', rotation=90, labelpad=15)
            
            # --- Cleaning up Axis and Adding Grid Lines ---
            ax.tick_params(left=False, right=False, labelleft=False, labelbottom=False)
            ax.axis('on')
            for spine in ax.spines.values():
                spine.set_edgecolor('gray')
                spine.set_linewidth(1.0)
            ax.set_xticks([]); ax.set_yticks([])

    plt.show()
    
# Example usage (assuming 'results_store' is available in the environment):
# plot_scanner_comparison(results_store)
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from typing import Dict, Tuple

def plot_tumor_type_comparison(
    results_store: Dict[str, Tuple[np.ndarray, int]],
    target_scanner: str = 'Hamamatsu XR',
    target_origin: str = 'UMC Utrecht'
):
    """
    Plots a simple row comparison of average images for a specific Scanner/Origin 
    pair across all available Tumor Types, separated by Mitotic/Non-Mitotic class.
    
    Args:
        results_store (Dict): Dictionary containing keys like 'Origin | Tumor | Scanner | Class'.
        target_scanner (str): The specific scanner to filter for.
        target_origin (str): The specific origin to filter for.
    """
    
    # --- 1. Filter and Prepare Data ---
    
    target_pattern = f"{target_origin} | " 
    target_suffix = f" | {target_scanner} | " 
    
    filtered_data = {}
    
    for key, (image, count) in results_store.items():
        if key.startswith(target_pattern) and target_suffix in key and image is not None and count > 0:
            parts = key.split(' | ')
            tumor_type = parts[1] 
            class_type = parts[3]
            
            if tumor_type not in filtered_data:
                filtered_data[tumor_type] = {}
            filtered_data[tumor_type][class_type] = {'Image': image, 'Count': count}
            
    if not filtered_data:
        print(f"Error: No data found for Scanner: '{target_scanner}' and Origin: '{target_origin}'.")
        return

    # --- 2. Setup Plot Dimensions ---

    unique_tumors = sorted(filtered_data.keys())
    n_tumors = len(unique_tumors)
    
    fig_width = n_tumors * 3.5
    fig_height = 2.5 * 2 + 1 
    
    fig, axes = plt.subplots(
        2, n_tumors, 
        figsize=(fig_width, fig_height), 
        squeeze=False, 
        # Hspace is for rows, wspace for columns. We rely on tight_layout for top margin.
        gridspec_kw={'wspace': 0.1, 'hspace': 0.4} 
    )
    
    plt.suptitle(f"Tumor Type Comparison for: {target_scanner} at {target_origin}", 
                 fontsize=16, y=0.98)

    # --- 3. Plotting Loop ---
    
    class_map = {'Mitotic': 0, 'Non-Mitotic': 1} 
    
    for c, tumor_type in enumerate(unique_tumors):
        tumor_data = filtered_data[tumor_type]
        
        # --- Column Title (Tumor Type Name) ---
        # FIX: Use set_title for automatic spacing via tight_layout
        axes[0, c].set_title(tumor_type, fontsize=12, fontweight='bold', pad=10)
        
        for class_type, row_idx in class_map.items():
            ax = axes[row_idx, c]
            item = tumor_data.get(class_type)
            
            if item:
                image = item['Image']
                count = item['Count']
                
                ax.imshow(image)
                
                # Label above the image
                color = 'darkred' if class_type == 'Mitotic' else 'darkblue'
                label = f"{class_type[0]} ({count})"
                # Placing label just outside the top border of the image
                ax.text(0.5, 1.15, label, transform=ax.transAxes, 
                        ha='center', fontsize=10, color=color, va='bottom')
            else:
                ax.text(0.5, 0.5, "No Data", ha='center', fontsize=8)

            # --- Row Label (Left Side) ---
            if c == 0:
                ax.set_ylabel(class_type, fontsize=12, fontweight='bold', rotation=90, labelpad=15)
            
            # --- Cleaning up Axis and Adding Grid Lines ---
            ax.tick_params(left=False, right=False, labelleft=False, labelbottom=False)
            ax.axis('on')
            for spine in ax.spines.values():
                spine.set_edgecolor('gray')
                spine.set_linewidth(1.0)
            ax.set_xticks([]); ax.set_yticks([])

    # FIX: Call tight_layout without rect to let it calculate the top margin for the titles
    plt.tight_layout()
    plt.show()
```

```python
plot_tumor_type_comparison(stored_averages)
```

Problem formulation:

1. Baseline, classification task: Insights training a CNN to identify mitotic or non-mitotic figures within a single scanner + origin + tumor dataset. canine lymphoma + 3D Histech + VMU Vienna
2. Shifts across scanner types: human breast cancer + UMC Utrecht
3. Shifts across origin type when target distribution labels are unavailable and limited: 3D histech + canine soft tissue sarcoma
4. Shifts across tumor type in combination with label shift: Hamamatsu XR + UMC Utrecht
5. Shifts across origin type when same-tumor/scanner training data is limited: All VMU Vienna + AMC New York
6. Combining to shift across tumor, scanner, and origin type and label shifts: canine cutaneous mast cell tumor | Aperio CS2 | FU Berlin vs all

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches 
from typing import List, Dict, Tuple

# Assume df_merged is already loaded and processed as in the previous steps.
# Placeholder data defined for reproducibility:
try:
    df_merged['category_id']
except NameError:
    data = {
        'Tumor': ['canine lymphoma', 'human breast cancer', 'canine soft tissue sarcoma', 'canine cutaneous mast cell tumor', 'human melanoma', 'human breast cancer', 'canine soft tissue sarcoma', 'canine lymphoma'],
        'Scanner': ['3D Histech', 'Hamamatsu XR', '3D Histech', 'Aperio CS2', 'Hamamatsu XR', 'Hamamatsu XR', '3D Histech', '3D Histech'],
        'Origin': ['VMU Vienna', 'UMC Utrecht', 'AMC New York', 'FU Berlin', 'UMC Utrecht', 'UMC Utrecht', 'VMU Vienna', 'UMC Utrecht'],
        'category_id': [1, 2, 1, 2, 1, 2, 2, 1] * 10, 
    }
    df_merged = pd.DataFrame(data)

# --- Experiment Mappings and Colors ---
EXPERIMENT_CRITERIA = {
    1: [('Tumor', 'canine lymphoma'), ('Scanner', '3D Histech'), ('Origin', 'VMU Vienna')],
    2: [('Tumor', 'human breast cancer'), ('Origin', 'UMC Utrecht')],
    3: [('Scanner', '3D Histech'), ('Tumor', 'canine soft tissue sarcoma')],
    4: [('Scanner', 'Hamamatsu XR'), ('Origin', 'UMC Utrecht')],
    5: [('Origin', 'VMU Vienna')],
    6: [('Tumor', 'canine cutaneous mast cell tumor'), ('Scanner', 'Aperio CS2'), ('Origin', 'FU Berlin')],
}

COLOR_MAP = {
    1: '#FFCCCB', 2: '#ADD8E6', 3: '#90EE90', 6: '#FFDAB9',
}

CONCISE_LABELS = {
    1: 'Baseline', 2: 'Covariate Scanner Shifts', 3: 'Covariate Origin Shifts under Limited Target Distribution Labels',
    4: 'Covariate Tumor Shifts with Label Shifts', 5: 'Handling Origin Shifts under Limited Same Scanner/Tumor Data',
    6: 'Final Experiment to Unseen Tumor, Scanner, Origin',
}

def map_row_to_experiments(row: pd.Series, criteria: Dict[int, List[Tuple[str, str]]]) -> List[int]:
    """Maps a table row to the list of experiments it belongs to."""
    experiments = []
    for exp_num, conditions_list in criteria.items():
        is_match = True
        for col, value in conditions_list:
            if row[col] != value:
                is_match = False
                break
        if is_match:
            experiments.append(exp_num)
    return experiments

# --- Table Data Preparation (Unchanged) ---
label_mapping = {2: 'Mitotic', 1: 'Non-Mitotic'}
df_merged['Mitotic_Class'] = df_merged['category_id'].map(label_mapping).fillna('Other')
df_counts = df_merged.groupby(['Tumor', 'Scanner', 'Origin', 'Mitotic_Class']).size().unstack(fill_value=0)
mitotic_counts = df_counts.get('Mitotic', 0)
non_mitotic_counts = df_counts.get('Non-Mitotic', 0)
df_counts['Total Annotations'] = mitotic_counts + non_mitotic_counts
df_counts['Mitotic Ratio'] = mitotic_counts / df_counts['Total Annotations']
df_counts['Non-Mitotic Ratio'] = non_mitotic_counts / df_counts['Total Annotations']
df_final_table = df_counts[['Total Annotations', 'Mitotic', 'Non-Mitotic', 'Mitotic Ratio', 'Non-Mitotic Ratio']].copy()
df_final_table['Mitotic Ratio'] = (df_final_table['Mitotic Ratio'] * 100).map('{:.2f}%'.format)
df_final_table['Non-Mitotic Ratio'] = (df_final_table['Non-Mitotic Ratio'] * 100).map('{:.2f}%'.format)
df_final_table = df_final_table.reset_index()
df_final_table = df_final_table.rename(columns={'Mitotic': 'Mitotic Annotations', 'Non-Mitotic': 'Non-Mitotic Annotations'})
df_final_table['Experiments'] = df_final_table.apply(lambda row: map_row_to_experiments(row, EXPERIMENT_CRITERIA), axis=1)
df_final_table['Experiments Involved'] = df_final_table['Experiments'].apply(
    lambda x: ', '.join(map(str, sorted(x))) if x else '-'
)

# --- Matplotlib Table Rendering with Coloring and Targeted Hatching ---
fig, ax = plt.subplots(figsize=(18, 7))
ax.axis('tight')
ax.axis('off')

table_data = df_final_table.drop(columns=['Experiments'])
cell_text = table_data.values
col_labels = table_data.columns

num_data_rows = len(df_final_table)
num_cols = len(col_labels)

cell_colors = [['w'] * num_cols for _ in range(num_data_rows)]
cell_hatches = [[''] * num_cols for _ in range(num_data_rows)] 

# Apply color (1, 2, 3, 6) and hatching (4, 5)
for i in range(num_data_rows):
    exp_list = df_final_table.loc[i, 'Experiments']
    
    # 1. Determine Base Color (Lowest of 1, 2, 3, 6)
    primary_exps = [e for e in exp_list if e in [1, 2, 3, 6]]
    if primary_exps:
        color_exp = min(primary_exps)
        color = COLOR_MAP[color_exp] 
    else:
        # If a row is ONLY 4 and/or 5, it gets a white background.
        color = 'w' 
        
    cell_colors[i] = [color] * num_cols
    
    # 2. Determine Hatching (Based on 4 and 5)
    hatch_str = ""
    if 4 in exp_list:
        hatch_str += '/' 
    if 5 in exp_list:
        hatch_str += '\\' 
        
    if hatch_str:
        cell_hatches[i] = [hatch_str] * num_cols

# Create the table
table = ax.table(
    cellText=cell_text, 
    colLabels=col_labels, 
    cellColours=cell_colors,
    cellLoc='center', 
    loc='center'
)

table.auto_set_font_size(False)
table.set_fontsize(8)
table.scale(1.2, 1.5) 

# Apply hatching with +1 row offset for the header
for i in range(num_data_rows):
    if cell_hatches[i][0] != '':
        for j in range(num_cols):
            cell_key = (i + 1, j) 
            if cell_key in table.get_celld():
                cell = table.get_celld()[cell_key]
                cell.set_hatch(cell_hatches[i][j])
                cell.set_edgecolor('black') 

# --- Add Title and Legend (Consolidated) ---
ax.set_title('Experiment Summary Table', y=0.98, fontsize=16, fontweight='bold') 

legend_handles = []
legend_labels = []

# Consolidated List of Experiments (Colors first, then Hatches)
# 1. Experiments with Colors (1, 2, 3, 6)
for exp_num in [1, 2, 3, 6]:
    color = COLOR_MAP[exp_num]
    label = f'Exp {exp_num} (Color): {CONCISE_LABELS[exp_num]}'
    patch = mpatches.Rectangle((0, 0), 1, 1, facecolor=color, edgecolor='black', linewidth=0.5)
    legend_handles.append(patch)
    legend_labels.append(label)

# 2. Experiments with Hatches (4, 5)

# Exp 4 Hatch
hatch_4 = mpatches.Rectangle((0, 0), 1, 1, facecolor='w', hatch='/', edgecolor='black', linewidth=0.5)
legend_handles.append(hatch_4)
legend_labels.append(f'Exp 4 (Hatch /): {CONCISE_LABELS[4]}')

# Exp 5 Hatch
hatch_5 = mpatches.Rectangle((0, 0), 1, 1, facecolor='w', hatch='\\', edgecolor='black', linewidth=0.5)
legend_handles.append(hatch_5)
legend_labels.append(f'Exp 5 (Hatch \\): {CONCISE_LABELS[5]}')

# Combined Hatch
hatch_45 = mpatches.Rectangle((0, 0), 1, 1, facecolor='w', hatch='\\/', edgecolor='black', linewidth=0.5)
legend_handles.append(hatch_45)

# Anchor the legend to a safe position below the table
ax.legend(
    legend_handles, 
    legend_labels, 
    title='Experiment Legend:',
    loc='lower left', 
    bbox_to_anchor=(0.0, -0.15), 
    fontsize=9,
    title_fontsize=10,
    frameon=False,
    ncol=1
)


plt.savefig('category_id_summary_table_final_legend.png', bbox_inches='tight')
plt.show()
```
